```python
class Celular:
    def __init__(self, marca, modelo, precio):
        self.marca = marca
        self.modelo = modelo
        self.precio = precio
        self.encendido = False

    def encender(self):
        self.encendido = True
        print(f"El {self.marca} {self.modelo} está encendido.")

    def mostrar_info(self):
        print(f"celular: {self.marca} {self.modelo}, precio $ {self.precio}")

# creacion de la clase
mi_celular1 = Celular("Samsung", "Galaxy S23", 800)
mi_celular2 = Celular("apple", "iphone 15", 1000)
# Llamada a las funciones para cada objeto celular 1 y celular 2
print("celular 1")
mi_celular1.mostrar_info()
mi_celular1.encender()

print("celular 2")
mi_celular2.mostrar_info()
mi_celular2.encender()

print("programa terminado p")
print("Creador de software por Christian GArcia NL = 21")
```
